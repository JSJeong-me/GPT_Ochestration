```python
import matplotlib.pyplot as plt
import numpy as np

# Create a 2D grid with 1cm spacing
x = np.arange(0, 10, 1)  # 10 arbitrary units along x-axis
y = np.arange(0, 10, 1)  # 10 arbitrary units along y-axis
X, Y = np.meshgrid(x, y)

# Plotting the grid
plt.figure(figsize=(8, 8))
plt.plot(X, Y, 'k-', lw=0.5)  # plot horizontal lines
plt.plot(Y, X, 'k-', lw=0.5)  # plot vertical lines
plt.title("2D Grid")
plt.xlabel("Fruit")
plt.ylabel("Ship")

# Display the grid
plt.show()
```
